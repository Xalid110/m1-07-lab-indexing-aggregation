```python
student_averages = scores.groupby(['cohort', 'student_id'])['score'].mean().reset_index()
student_averages.rename(columns={'score': 'avg_score'}, inplace=True)
student_averages['rank'] = student_averages.groupby('cohort')['avg_score'].rank(
    method='min',
    ascending=False
)
top_students = student_averages[student_averages['rank'] <= 3][['student_id', 'cohort', 'avg_score']]
```

```python
top_students
```

```python
cohort_counts = top_students['cohort'].value_counts()
is_valid = (cohort_counts == 3).all()
is_valid#check cohort appear 3 times
```
